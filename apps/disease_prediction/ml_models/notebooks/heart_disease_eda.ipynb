```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
```

```python
DATA_PATH = "../datasets/heart_disease_train.csv"

df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Numerical columns
num_cols = [
    "Age", "RestingBP", "Cholesterol", "MaxHR", "Oldpeak"
]

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns
cat_cols = [
    "Sex", "ChestPainType", "RestingECG",
    "ExerciseAngina", "ST_Slope"
]

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isnull().sum()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="HeartDisease")
plt.title("Target Distribution")
plt.show()
```

```python
sns.countplot(x="Sex", hue="HeartDisease", data=df)
plt.title("Heart Disease by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="ChestPainType", hue="HeartDisease", data=df)
plt.title("Chest Pain Type vs Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="HeartDisease", y="Cholesterol", data=df)
plt.title("Cholesterol vs Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.head()
```

```python
print("Age Statistics")
print("-" * 40)

print("Min Age:", df["Age"].min())
print("Max Age:", df["Age"].max())
print("Mean Age:", round(df["Age"].mean(), 2))
print("Median Age:", df["Age"].median())
print("Std Dev Age:", round(df["Age"].std(), 2))
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=25, kde=True)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr_target = corr['HeartDisease']

plt.figure(figsize=(8, 4))

# Drop target itself
corr_target.drop('HeartDisease').plot(kind='bar')

plt.title("Feature Correlation with Heart Disease")
plt.ylabel("Correlation Value")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
